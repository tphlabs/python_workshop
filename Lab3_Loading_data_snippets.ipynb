# Loading handwritten data <a name="hand"></a>

Can be used for small number of datapoints.

```python
measurements = '''
#	time[s]	x [cm]
0.0050 54
0.0100 52
0.0150 48
0.0200 46
0.0250 42
0.0300 39
0.0350 36
0.0400 33
0.0450 30
0.0500 26
'''
t, x = np.genfromtxt(measurements.splitlines()).T
plt.plot(t, x, '.')
plt.show()
```

# Loading data from Kruze sensor <a name="kruze"></a>



Sample data file with counters B and C

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/main/Lab3_CO/Experiment_2023.11.26/2dof/free_C.txt'

_, t, countB, countC = np.genfromtxt(filename).T
plt.plot(t, countB, label='B')
plt.plot(t, countC, label='C')
plt.legend()
plt.show()
```

Sample data file with one counter C

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/main/Lab3_CO/Experiment_2023.11.26/1dof/free_1.txt'
_, t, countC = np.genfromtxt(filename).T
plt.plot(t, countC, label='C')
plt.legend()
plt.show()
```

# Loading data from Keysight digital scope <a name="scope"></a>

Two channels data.




```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/main/Lab2_R/2024_05_15/cuprum_1.csv'
t, ch1, ch2 = np.genfromtxt(filename,
                            delimiter=',',
                            usecols=[1,2,3],
                            skip_header=47).T
plt.plot(t, ch1, label='Channel 1, [V]')
plt.plot(t, ch2, label='Channel 2, [V]')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend()
plt.show()
```

One channel data.

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/main/Lab2_R/2024_05_15/cuprum_1.csv'
t, ch1 = np.genfromtxt(filename,
                            delimiter=',',
                            usecols=[1,2],
                            skip_header=47).T
plt.plot(t, ch1, label='Channel 1, [V]')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend()
plt.show()
```
