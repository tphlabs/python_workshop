```python
import numpy as np # working with vectors, matrices and loading data
import matplotlib.pyplot as plt # plotting data
```

# Loading handwritten data <a name="hand"></a>

Can be used for small number of datapoints. Data is kept in multiline string.

```python
measurements = '''
#	time[s]	x [cm]
0.05 54
0.10 52
0.15 48
0.20 46
0.25 42
0.30 39
0.35 36
0.40 33
0.45 30
0.50 26
'''
t, x = np.genfromtxt(measurements.splitlines()).T
plt.plot(t, x, '.')
plt.xlabel('Time [s]')
plt.ylabel('x [cm]')
plt.show()
```

# Loading data from Kruze sensor <a name="kruze"></a>

Space separated values (default separator).

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/main/Lab3_CO/Experiment_2023.11.26/2dof/free_C.txt'

_, t, countB, countC = np.genfromtxt(filename).T
plt.plot(t, countB, label='B')
plt.plot(t, countC, label='C')
plt.legend()
plt.show()
```

# Loading data from Keysight digital scope <a name="scope"></a>

`CSV` - comma separated values.

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/main/Lab2_R/2024_05_15/cuprum_1.csv'
t, ch1, ch2 = np.genfromtxt(filename,
                            delimiter=',',
                            usecols=[1,2,3],
                            skip_header=47).T
plt.plot(t, ch1, label='Channel 1, [V]')
plt.plot(t, ch2, label='Channel 2, [V]')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend()
plt.show()
```

# Loading data from Geiger counter <a name="geiger"></a>

`TSV` - tab separated values.

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/main/Lab4_Nuclear/2023_12_13/Co60_shelf10_101measurements.tsv'

counts, time = np.genfromtxt(filename,
                            delimiter='\t',
                            usecols=[2,3],
                            skip_header=11).T
intensity = counts / time
plt.hist(intensity, density=True);
plt.xlabel('Intensity, 1/s')
plt.ylabel('Probability')
plt.show()
```

# Loading data from MCA <a name="mca"></a>

`TXT`with tab delimiter

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/refs/heads/main/Lab4_Xray/2024.10.30/Mo_calibration.txt'

channel, intensity = np.genfromtxt(filename,
                            delimiter='\t',
                            skip_header=3).T

plt.figure(figsize=(12, 4))
plt.plot(channel, intensity, '.', color='gray')
plt.xlabel('Channel')
plt.ylabel('Intensity')
plt.show()
```
